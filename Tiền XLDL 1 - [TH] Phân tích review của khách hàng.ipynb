```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler
```

```python
df = pd.read_csv("D:\DA-Class\DA2212R1\Credit_Scoring.csv")
df.head()
```

```python
# thông tin dữ liệu
df.info()
```

```python
# mô tả dữ liệu
df.describe()
```

```python
# kiểm tra dữ liệu khuyết thiếu
df.isna()
```

```python
# loại bỏ dữ liệu khuyết thiếu
df1 = df.dropna()
```

```python
# % số lượng bản ghi còn lại
100 * df1.shape[0]/df.shape[0]
```

```python
# vẽ biểu đồ phân bố
sns.kdeplot(data=df1["MonthlyIncome"])
```

Các đặc trưng đều là giá trị liên tục nên ta sẽ áp dụng phương pháp thay thế nội suy

```python
# thay thế dữ liệu khuyết thiếu bởi giá trị nội suy theo cột
df2 = df.interpolate(axis=1)
```

```python
df2.isna()
```

```python
# vẽ biểu đồ boxplot cho các đặc trưng
df2.boxplot()
```

```python
# vẽ biểu đồ box plot cho MonthlyIncome
sns.boxplot(df2["MonthlyIncome"])
```

```python
# tính giá trị Q1 và Q3
Q1 = df2.quantile(0.25)
Q3 = df2.quantile(0.75)
```

```python
# tính IQR 
IQR = Q3-Q1
```

```python
# lọc dữ liệu ngoại lai
df3 = df2[~((df2 < (Q1 - 1.5 * IQR)) | (df2 > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df3.boxplot()
```

```python
sns.boxplot(df3["MonthlyIncome"])
```

```python
df3.describe()
```

```python
# phân bố dữ liệu trên cột MonthlyIncome
sns.kdeplot(data = df3['MonthlyIncome'])
```

```python
# chuẩn hóa với minmax scaling
scaler = MinMaxScaler()

mms = scaler.fit_transform(pd.DataFrame(df3['MonthlyIncome']))
```

```python
sns.kdeplot(data = mms)
```

Có thể thấy phân bố dữ liệu vẫn được giữ nguyên, tuy nhiên khoảng giữ liệu đã được thu hẹp về 0-1

```python
# chuẩn hóa với robust scaling
scaler = RobustScaler()

rbs = scaler.fit_transform(pd.DataFrame(df3['MonthlyIncome']))
```

```python
sns.kdeplot(data = rbs)
```

```python
# chuẩn hóa với standard scaling
scaler = StandardScaler()

sc = scaler.fit_transform(pd.DataFrame(df3['MonthlyIncome']))
```

```python
sns.kdeplot(data = sc)
```

Có thể nhận thấy ở cả 3 phương pháp chuẩn hóa, phân bố dữ liệu gần như được giữ nguyên, tuy nhiên khoảng giá trị giữ liệu đã thay đổi về khoảng giá trị nhỏ hơn rất nhiều so với ban đầu
