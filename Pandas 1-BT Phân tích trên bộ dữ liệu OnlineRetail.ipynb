```python
import pandas as pd
import numpy as np
```

```python
#1.Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
import pandas as pd
df = pd.read_csv('Datasets\OnlineRetail.csv', encoding = "ISO-8859-1")
df
```

```python
print("Số dòng của dữ liệu là: ", df.shape[0])
print("Số cột của dữ liệu là: ", df.shape[1])
df.info()
```

```python
#2. Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv
df2 = df.loc[:,['Description','Quantity']]
df2
```

```python
df2.to_csv('OnlineRetail_BT.csv')
```

```python
# 3.Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx
df3 = df.iloc[0:1000]
df3
```

```python
df3.to_excel('OnlineRetail.xlsx')
```

```python
# 4.Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5
df4 = df.loc[df.Quantity >= 10]
df4
```

```python
df.to_hdf('OnlineRetail.h5', 'table')
```

```python
# 5.Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, 
# các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json
df5 = df.iloc[1000:2001].loc[:,['Quantity','InvoiceDate','UnitPrice']]
df5
```

```python
df5.to_json('OnlineRetail.json',orient='columns')
```

```python
# 6.Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html
df6 = df.loc[df.InvoiceNo == '536365']
df6
```

```python
df6.to_html('OnlineRetail.html')
```
